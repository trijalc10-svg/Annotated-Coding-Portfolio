```python
## a to do list manager which uses functions, lists for storing tasks, file handling and error handling
def save_task(tasks): # defines function save_tasks which takes input of tasks
  with open("to do.txt", "w") as f: # with closes the file when done with it, open file named to do.txt in write mode
    for task in tasks: # for every task in the tasks list
      f.write(task + "\n") # write in the file, the name of the task in seperate lines

try: # try to run following code
    with open("to do.txt", "r") as f: # open this file in read mode and close it once finished
      tasks = [line.strip() for line in f] # variable tasks stored as a list, all new lines and spaces in the file removed to put it into a list
except FileNotFoundError: # catches this error, if the file doesn't exist yet
  with open("to do.txt", "w") as f: # open the file in write mode
    pass # allows the file to be created
    tasks = [] # task stored as empty list

while True: # while true makes the loop always play, every time user does something this menu is deplayed again as they're sent to the start
  print("\n1) view tasks") # creates new line and then prints the options 1-5
  print("2) add tasks ")
  print("3) mark as complete")
  print("4) delete task")
  print("5) exit")
  choice = int(input("Choose an option 1-5: ")) # accepts input from the user as an integer to choose


  valid = (choice >= 1 and choice <= 5) # boolean logic to see if the choice is valid

  if not valid: # if invalid, tell the user
    print("invalid choice") # end of loop if invalid so user gets sent to the start of the loop
  else: # otherwise
    if choice == 1: # they choose to view tasks
      if not tasks: # means if tasks is false, meaning the list is empty
        print("You have no tasks") # tell user they have no tasks
      else:# otherwise
        for task in tasks: # for every task in the tasks list
          print("\n" + task) # print each task on a seperate line
    elif choice == 2: # user wants to add tasks
      new_task = input("task name: ") # accepts input and stores in new_task variable
      tasks.append(new_task) # user input is added as a new item at the end of a list
      print(f"{new_task} added") # tells the user the name of the new task and that it has been added
      save_task(tasks) # saves new tasks list to the to do.txt file
    elif choice == 3: # if the user wants to mark a task as complete
      if not tasks: # if the tasks list is empty
        print("You have no tasks to mark") # tell user there's nothing to mark
      else: # otherwise
        for i, task in enumerate(tasks, 1): # each task in tasks is assigned a number starting from 1 to display, i is the number, task is the task name
          print(f"\n{i}) {task}") # print the task numbers with their names
        mark = int(input("Enter number of task you want to complete: ")) - 1 # ask user to input integer - 1 as indexing starts at 0 and our list displayed starts at 1
        if 0 <= mark < len(tasks): # if the mark is within the length of the list
          tasks[mark] += " - Complete" # take the item of the list with the corresponding number and add complete to the end
          print("marked as complete") # tell the user it has been done
          save_task(tasks) # save new list to file
        else: # otherwise
          print("invalid task") # tell user task doesn't exist
    elif choice == 4: # if they want to delete task
      if not tasks: # if there are no tasks
        print("You have no tasks to delete") # tell user
      else: # otherwise
        for i, task in enumerate(tasks, 1): # assigns each task a number starting from 1 to display
          print(f"\n{i}){task}") # display task with number
        delete = int(input("Enter number of task you want to delete: ")) - 1 # accept integer input and minus 1 since indexing starts at 0
        if 0 <= delete < len(tasks): # if input is within length of list
          deleted_task = tasks.pop(delete) # delete the task corresponding to the number and stoer the task name as deleted_task
          save_task(tasks) # save new tasks list to file
          print(f"{deleted_task} has been deleted") # tell the user what task has been deleted
        else: # otherwise
            print("invalid task number") # tell the user the task doesn't exist

    elif choice == 5: # if they want to exit program
        print("Exiting program.") # tell them they are exiting
        break # stop code running all together
```
